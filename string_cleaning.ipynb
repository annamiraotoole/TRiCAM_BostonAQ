```python
def change_names(land_use_column):
    for i in range(len(land_use_column)):
        mydict = yaml.load(land_use_column[i])
        if 'Commercial and Services' in mydict:
            mydict['commercial'] = mydict.pop('Commercial and Services')
            
        if 'Transportation, communications and services' in mydict:
            mydict['transportation'] = mydict.pop('Transportation, communications and services')
        
        if 'Other urban or built-up land' in mydict:
            mydict['urban_public/institution'] = mydict.pop('Other urban or built-up land')
        
        if 'Cropland and pasture' in mydict:
            mydict['crop_land'] = mydict.pop('Cropland and pasture')
        
        if 'Orchards, groves, vineyards, nurseries, and ornamental horticultural' in mydict:
            mydict['crop_land'] = mydict.pop('Orchards, groves, vineyards, nurseries, and ornamental horticultural')
       
        if 'Confined feeding operations' in mydict:
            mydict['crop_land'] = mydict.pop('Confined feeding operations')
       
        if 'Other agricultural land' in mydict:
            mydict['crop_land'] = mydict.pop('Other agricultural land')
        
        if 'Herbaceous rangeland' in mydict:
            mydict['open_land'] = mydict.pop('Herbaceous rangeland')
        
        if 'Shrub and brush rangeland' in mydict:
            mydict['open_land'] = mydict.pop('Shrub and brush rangeland')
        
        if 'Mixed rangeland' in mydict:
            mydict['open_land'] = mydict.pop('Mixed rangeland')
        
        if 'Deciduous forest land' in mydict:
            mydict['forest'] = mydict.pop('Deciduous forest land')
            
        if 'Evergreen forest land' in mydict:
            mydict['forest'] = mydict.pop('Evergreen forest land')
            
        if 'Mixed forest land' in mydict:
            mydict['forest'] = mydict.pop('Mixed forest land')
        
        if 'Streams and canals' in mydict:
            mydict['water'] = mydict.pop('Streams and canals')
            
        if 'Lakes' in mydict:
            mydict['water'] = mydict.pop('Lakes')
            
        if 'Reservoirs' in mydict:
            mydict['water'] = mydict.pop('Reservoirs')
            
        if 'Bays and estuaries' in mydict:
            mydict['water'] = mydict.pop('Bays and estuaries')
            
        if 'Forested wetland' in mydict:
            mydict['wetland'] = mydict.pop('Forested wetland')
            
        if 'Nonforested wetland' in mydict:
            mydict['wetland'] = mydict.pop('Nonforested wetland')
            
        if 'Dry salt flats' in mydict:
            mydict['open_land'] = mydict.pop('Dry salt flats')
            
        if 'Bare exposed rock' in mydict:
            mydict['open_land'] = mydict.pop('Bare exposed rock')
            
        if 'Mixed Barren Land' in mydict:
            mydict['open_land'] = mydict.pop('Mixed Barren Land')
            
        if 'Beaches' in mydict:
            mydict['saltwater/sandybeach'] = mydict.pop('Beaches')
    
        if 'Sandy areas not beaches' in mydict:
            mydict['saltwater/sandybeach'] = mydict.pop('Sandy areas not beaches')
            
        if 'Strip mines, quarries, gravel pits' in mydict:
            mydict['mining'] = mydict.pop('Strip mines, quarries, gravel pits')
            
        if 'Transitional areas' in mydict:
            mydict['transitional'] = mydict.pop('Transitional areas')
            
        if 'Shrub and brush tundra' in mydict:
            mydict['open_space'] = mydict.pop('Shrub and brush tundra')
            
        if 'Herbaceous tundra' in mydict:
            mydict['open_space'] = mydict.pop('Herbaceous tundra')
            
        if 'Bare ground' in mydict:
            mydict['open_space'] = mydict.pop('Bare ground')
            
        if 'Wet tundra' in mydict:
            mydict['open_space'] = mydict.pop('Wet tundra') 
            
        if 'Mixed tundra' in mydict:
            mydict['open_space'] = mydict.pop('Mixed tundra')
            
        if 'Perennial snowfields' in mydict:
            mydict['open_space'] = mydict.pop('Perennial snowfields')
            
        if 'Glaciers' in mydict:
            mydict['open_space'] = mydict.pop('Glaciers')
            
        if 'Industrial' in mydict:
            mydict['industrial'] = mydict.pop('Industrial')
            
        if 'Residential' in mydict:
            mydict['residential'] = mydict.pop('Residential')

        land_use_column.set_value(i, mydict)
    return land_use_column
```

```python
def update_proportions(landuse_col, df):
    for i in range(len(landuse_col)):
        line = json.dumps(landuse_col[i])
        my_dict = json.loads(line)
        for j in my_dict:
            if 'forest' in my_dict:
                df['forest'] = np.where(df["forest"] == i, df['forest']+my_dict.get('forest'), df['forest'])

            if 'open_land' in my_dict:
                df['open_land'] = np.where(df['open_land'] == i, df['open_land']+my_dict.get('open_land'), df['open_land'])

            if 'water' in my_dict:
                df['water'] = np.where(df["water"] == i, df['water']+my_dict.get('water'), df['water'])

            if 'wetland' in my_dict:
                df['wetland'] = np.where(df["wetland"] == i, df['wetland']+my_dict.get('wetland'), df['wetland'])

            if 'saltwater/sandybeach' in my_dict:
                df['saltwater/sandybeach'] = np.where(df["saltwater/sandybeach"] == i, df['saltwater/sandybeach']+my_dict.get('saltwater/sandybeach'), df['saltwater/sandybeach'])

            if 'recreational' in my_dict:
                df['recreational'] = np.where(df["recreational"] == i, df['recreational']+my_dict.get('recreational'), df['recreational'])

            if 'marina' in my_dict:
                df['marina'] = np.where(df["marina"] == i, df['marina']+my_dict.get('marina'), df['marina'])

            if 'transitional' in my_dict:
                df['transitional'] = np.where(df["transitional"] == i, df['transitional']+my_dict.get('transitional'), df['transitional'])

            if 'urban_public/institution' in my_dict:
                df['urban_public/institution'] = np.where(df["urban_public/institution"] == i, df['urban_public/institution']+my_dict.get('urban_public/institution'), df['urban_public/institution'])

            if "commercial" in my_dict:
                df['commercial'] = np.where(df["commercial"] == i, df['commercial']+my_dict.get('commercial'), df['commercial'])

            if "transportation" in my_dict:
                df['transportation'] = np.where(df["transportation"] == i, df['transportation']+my_dict.get('transportation'), df['transportation'])

            if 'mining' in my_dict:
                df['mining'] = np.where(df["mining"] == i, df['mining']+my_dict.get('mining'), df['mining'])

            if 'waste' in my_dict:
                df['waste'] = np.where(df["waste"] == i, df['waste']+my_dict.get('waste'), df['waste'])

            if 'crop_land' in my_dict:
                df['crop_land'] = np.where(df["crop_land"] == i, df['crop_land']+my_dict.get('crop_land'), df['crop_land'])

            if 'cemetary' in my_dict:
                df['cemetary'] = np.where(df["cemetary"] == i, df['cemetary']+my_dict.get('cemetary'), df['cemetary'])

            if 'golfcourse' in my_dict:
                df['golfcourse'] = np.where(df["golfcourse"] == i, df['golfcourse']+my_dict.get('golfcourse'), df['golfcourse'])

            if 'high_density_residential' in my_dict:
                df['forest'] = np.where(df["high_density_residential"] == i, df['high_density_residential']+my_dict.get('high_density_residential'), df['high_density_residential'])

            if 'medium_density_residential' in my_dict:
                df['medium_density_residential'] = np.where(df["medium_density_residential"] == i, df['medium_density_residential']+my_dict.get('medium_density_residential'), df['medium_density_residential'])

            if 'low_density_residential' in my_dict:
                df['low_density_residential'] = np.where(df["low_density_residential"] == i, df['low_density_residential']+my_dict.get('low_density_residential'), df['low_density_residential'])
                
            if 'industrial' in my_dict:
                df['industrial'] = np.where(df["industrial"] == i, df['industrial']+my_dict.get('industrial'), df['industrial'])
    return df
```
