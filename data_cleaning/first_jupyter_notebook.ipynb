```python
def greatest(list):
    answer = list[0]
    for n in list:
        if n > answer:
            answer = n
    return answer
```

```python
greatest([2, 3, 7, 6, 8, 34])
```

```python
def list_tricks(list):
    print(list[0:-1:2])
    print(list[::-1])
    print(list[-5:])
```

```python
list_tricks("abcdefghij")
```

```python
def palindrome(str):
    if str == str[::-1]:
        return True
    else:
        return False
```

```python
palindrome('hannah'), palindrome('hanna')
```
