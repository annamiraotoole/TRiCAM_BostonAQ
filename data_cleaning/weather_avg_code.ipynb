```python
import pandas as pd
import csv
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
```

```python
path = 'wunderground_csvs/'

BackBay = pd.read_csv(path + 'BackBay.csv')
Belmont = pd.read_csv(path + 'Belmont.csv')
CambridgePort = pd.read_csv(path + 'CambridgePort.csv')
ChestnutHill = pd.read_csv(path + 'ChestnutHill.csv')
Dorchester = pd.read_csv(path + 'Dorchester.csv')
# Logan = pd.read_csv(path + 'Logan.csv')
Malden = pd.read_csv(path + 'Malden.csv')
Medford = pd.read_csv(path + 'Medford.csv')
Waban = pd.read_csv(path + 'Waban.csv')

dataframes = [BackBay, Belmont, CambridgePort, ChestnutHill, Dorchester, Malden, Medford, Waban]
```

```python
for df in dataframes:
    
    del df['image']
    del df['display_location']
    del df['observation_location']
    del df['estimated']
    del df['observation_time_rfc822']
    del df['observation_epoch']
    del df['local_time_rfc822']
    del df['local_epoch']
    del df['local_tz_short']
    del df['local_tz_long']
    del df['local_tz_offset']
    del df['weather']
    del df['temperature_string']
    del df['wind_string']
    del df['dewpoint_string']
    del df['heat_index_string']
    del df['windchill_string']
    del df['feelslike_string']
    del df['precip_1hr_string']
    del df['precip_today_string']
    del df['icon']
    del df['icon_url']
    del df['forecast_url']
    del df['history_url']
    del df['ob_url']
    del df['nowcast']
    del df['wind_dir']
    del df['pressure_trend']
    del df['precip_today_metric']
    del df['solarradiation']
    del df['relative_humidity']
```

```python
for df in dataframes:
    
    for i in range(df.shape[0]):
        df['relative_humidity'][i] = int(df['relative_humidity'][i][:-1])
```

```python
# open new file for current station's data
file = open('wunderground_csvs/weather_averages.csv', 'w')

# create the csv writer object
csvwriter = csv.writer(file)

csvwriter.writerow(BackBay.columns)

for df in dataframes:
    
    avg = np.zeros(df.shape[1])
    columns = df.columns
    
    for i in range(len(columns)):
        
        avg[i] = df[columns[i]].mean()
    
    csvwriter.writerow(avg)
```
